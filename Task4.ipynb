```python
from sklearn import datasets
import pandas as pd
```

```python
iris=datasets.load_iris()
```

```python
x=pd.DataFrame(iris.data, columns=iris.feature_names)
x.head(5)
```

```python
y=iris.target
y
```

### Splitting model into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

### Dicision tree classifier

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier()
dtc.fit(x_train,y_train)
Predicted = dtc.predict(x_test)
```

```python
#Checking accuracy
from sklearn import metrics
print('Accurecy : ', metrics.accuracy_score(Predicted, y_test))
```

```python
from sklearn.externals.six import StringIO
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
from sklearn import tree
import matplotlib.image as img
import matplotlib.pyplot as plt
import graphviz
```

```python
dot_data = StringIO()

export_graphviz(dtc, out_file=dot_data, feature_names=iris.feature_names,  
                filled=True, rounded=True,
                special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
```

```python
graph.write_png("IrisTree.png")
DT=img.imread("IrisTree.png")
plt.figure(figsize=(15,30))
plt.imshow(DT, interpolation='nearest')
```
